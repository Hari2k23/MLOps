Logistic Regression

```python
#Importing necessary Python Libraries
import pandas as pd
import numpy as np
```

```python
#Importing the dataset
data = pd.read_csv("/content/Diabetes.csv")
data.head()
```

```python
data.describe()
```

```python
#Checking missing values
data.isna().sum()
```

```python
# Spliting the data into features and target
X = data.iloc[:,:-1]
y = data['Outcome']
```

```python
#Splitting the data into Training and Testing datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
#Training the Logistic Regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
#Making predictions
y_pred = model.predict(X_test)
```

```python
# Evaluating the model
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall Score:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
```

```python
import pickle
model_pkl_file = "/content/Diabetes.pkl"
with open(model_pkl_file, 'wb') as file:
    pickle.dump(model, file)

loaded_model = pickle.load(open('/content/Diabetes.pkl', 'rb'))

input_data = np.array([[6,	148.0,	72.0,	35.00000,	155.548223,	33.6,	0.627,	50]])

#input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = loaded_model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('Diabetes Negative')
if (prediction[0] == 1):
  print('Diabetes Positive')
```
